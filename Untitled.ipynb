```python
# Membuat list berisi data angka penjualan/nilai
data = [15, 30, 20, 18, 40]

# Inisialisasi variabel total dengan nilai awal 0
total = 0

# Melakukan perulangan untuk menjumlahkan setiap angka di dalam list data
for i in data:
    total += i
# Menampilkan hasil total penjumlahan angka
print(total)
```

```python
# Membuat fungsi untuk menghitung rata-rata (total jumlah dibagi banyak data)
def rata_rata(data) :
    return sum(data) / len(data)

# Membuat list angka baru
nilai = [15, 30, 20]
# Memanggil fungsi rata_rata dan menampilkan hasilnya
print(rata_rata(nilai))
```

```python
# Membuat blueprint/class bernama Produk
class Produk:
    # Konstruktor untuk inisialisasi properti nama dan jumlah produk
    def __init__(self, nama, jumlah):
        self.nama = nama
        self.jumlah = jumlah

    # Method untuk menampilkan data produk ke layar
    def tampil_data(self):
        print(self.nama, self.jumlah)

# Membuat objek p1 dari class Produk dengan nama 'Laptop' dan jumlah 15
p1 = Produk("Laptop", 15)
# Menjalankan method tampil_data untuk melihat hasilnya
p1.tampil_data()
```

```python
# Import library Pandas untuk manipulasi tabel/data dataframes
import pandas as pd
# Membaca file data eksternal berformat CSV
data = pd.read_csv("penjualan.csv")

# Menampilkan 5 baris data teratas dari file CSV
print(data.head())

# Menghitung dan menampilkan total keseluruhan dari kolom 'Penjualan'
print(data["Penjualan"].sum())
```

```python
# Import library Numpy untuk operasi matematika/array
import numpy as np

# Membuat array Numpy berisi kumpulan angka penjualan
penjualan = np.array([15,30,20,18,40])

# Menghitung dan menampilkan nilai rata-rata (mean), nilai tertinggi (max), dan terendah (min)
print(np.mean(penjualan))
print(np.max(penjualan))
print(np.min(penjualan))
```

```python
# Import library Matplotlib untuk visualisasi data grafik
import matplotlib.pyplot as plt

# Mengelompokkan data berdasarkan nama produk, lalu menjumlahkan penjualannya
produk = data.groupby("Produk")["Penjualan"].sum()

# Membuat grafik batang menggunakan data produk dan nilai penjualannya
plt.bar(produk.index, produk.values)
plt.title("Grafik Penjualan")
plt.xlabel("Produk")
plt.ylabel("Total Penjualan")

# Menampilkan grafik yang sudah dikonfigurasi ke layar
plt.show()
```
