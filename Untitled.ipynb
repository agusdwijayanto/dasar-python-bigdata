```python
def rata_rata(data) :
    return sum(data) / len(data)

nilai = [15, 30, 20]
print(rata_rata(nilai))
```

```python
class Produk:
    def __init__(self, nama, jumlah):
        self.nama = nama
        self.jumlah = jumlah

    def tampil_data(self):
        print(self.nama, self.jumlah)

p1 = Produk("Laptop", 15)
p1.tampil_data()
```

```python
import pandas as pd
data = pd.read_csv("penjualan.csv")

print(data.head())

print(data["Penjualan"].sum())
```

```python
import numpy as np

penjualan = np.array([15,30,20,18,40])

print(np.mean(penjualan))
print(np.max(penjualan))
print(np.min(penjualan))
```

```python
import matplotlib.pyplot as plt

produk = data.groupby("Produk")["Penjualan"].sum()

plt.bar(produk.index, produk.values)
plt.title("Grafik Penjualan")
plt.xlabel("Produk")
plt.ylabel("Total Penjualan")

# 4. Tampilkan grafik
plt.show()
```
